```python
# Loading Necessary Packages
import pandas as pd
import requests, shutil
import os
import time
```

```python
# Loading Category data
flipkart_cat = pd.read_csv("./Data/Flipkart_category.csv")
# Forming category data list
categor_list = flipkart_cat.apply(lambda x: x.dropna().tolist()).tolist()
# Filter data with regx
Shorts = '|'.join(categor_list[0])
Pants =  '|'.join(categor_list[1])
Mixed = '|'.join(categor_list[2])
Tops = '|'.join(categor_list[3])
Jackets = '|'.join(categor_list[4])
Jumpsuit = '|'.join(categor_list[5])
Bottom = '|'.join(categor_list[6])
Hoodie = '|'.join(categor_list[7])
# Generalized category assignment
my_data.loc[my_data['product_category_tree'].str.contains(Shorts),"category_type"] = "Shorts"
my_data.loc[my_data['product_category_tree'].str.contains(Pants),"category_type"] = "Pants"
my_data.loc[my_data['product_category_tree'].str.contains(Mixed),"category_type"] = "Mixed"
my_data.loc[my_data['product_category_tree'].str.contains(Tops),"category_type"] = "Tops"
my_data.loc[my_data['product_category_tree'].str.contains(Jackets),"category_type"] = "Jackets"
my_data.loc[my_data['product_category_tree'].str.contains(Bottom),"category_type"] = "Bottom"
my_data.loc[my_data['product_category_tree'].str.contains(Jumpsuit),"category_type"] = "Jumpsuit"
my_data.loc[my_data['product_category_tree'].str.contains(Hoodie),"category_type"] = "Hoodie"
# category type
my_data['category_type'].value_counts()
```

### Myntra Dataset

```python
#Myntra_Dataset
myntra = pd.read_csv("./Data/myntra_ecommerce.csv")
myntra_cat = pd.read_csv("./Data/myntra_category.csv")
# Dataset Shape
myntra.shape
```

```python
#First 5 observations
myntra.head()
```

```python
# Creating image lists
myntra['images'] = myntra['images'].str.split('|')
# Rows with unique image url
myntra = myntra.assign(url_img=myntra['images']).explode('url_img').reset_index(drop=True)
myntra.shape
```

```python
myntra.head()
```

```python
# Segmenting required columns
myntra_data = myntra[['link','product_type','uniq_id','images','url_img']].copy()
myntra_data['Web'] = "Myntra"
```

```python
# Category list
categor_list_myn = myntra_cat[['Dress','Tops', 'Mixed', 'Jacket', 'Pants', 'Bottom', 'Jumpsuit', 'Hoodie']].apply(lambda x: x.dropna().tolist()).tolist()
# Filter data with regx
Tops = '|'.join(categor_list_myn[1])
Mixed =  '|'.join(categor_list_myn[2])
Jacket = '|'.join(categor_list_myn[3])
Pants = '|'.join(categor_list_myn[4])
Bottom = '|'.join(categor_list_myn[5])
Jumpsuit = '|'.join(categor_list_myn[6])
Hoodie = '|'.join(categor_list_myn[7])
Dress = '|'.join(categor_list_myn[0])
# Droping nan products
myntra_data = myntra_data.dropna(subset=['product_type'])
# Generalized category assignment
myntra_data.loc[myntra_data['product_type'].str.contains(Pants),"category_type"] = "Pants"
myntra_data.loc[myntra_data['product_type'].str.contains(Mixed),"category_type"] = "Mixed"
myntra_data.loc[myntra_data['product_type'].str.contains(Tops),"category_type"] = "Tops"
myntra_data.loc[myntra_data['product_type'].str.contains(Jacket),"category_type"] = "Jackets"
myntra_data.loc[myntra_data['product_type'].str.contains(Bottom),"category_type"] = "Bottom"
myntra_data.loc[myntra_data['product_type'].str.contains(Jumpsuit),"category_type"] = "Jumpsuit"
myntra_data.loc[myntra_data['product_type'].str.contains(Hoodie),"category_type"] = "Hoodie"
myntra_data.loc[myntra_data['product_type'].str.contains(Dress),"category_type"] = "Dress"
# Category Type
myntra_data['category_type'].value_counts()
```

```python
# Merging Flipkart & Myntra Dataset
# Renaming columns to match both dataset
myntra_data = myntra_data.rename(columns={"link": "product_url", "product_type": "product_category_tree", "uniq_id":"pid","images":"image" })
# Conacatenating data
my_data = my_data.append(myntra_data, ignore_index=True)
# Removing nan datas
my_data = my_data.dropna(subset=['category_type'])
# Shape after adding
my_data.shape
```

```python
#Category Type
my_data['category_type'].value_counts()
```

### Image Data Download

```python
category = list(my_data['category_type'].unique())
# Defining filename to images
def generate_filename(loc):
    '''
    Each product ID has many images - images needs to named appropriately
    input: product details - Flipkart/Myntra and location details of the file
    ouput: New unique filename - Productid_<iteration value>
    '''
    if os.path.isfile(loc):
        name = loc.split('_')
        if len(name) == 2:
            loc = str(name[0])+'_'+str(int(name[1][:-4])+1)+'.jpg'
        else:
            loc = str(name[0][:-4])+'_1'+'.jpg'
        return generate_filename(loc)
    return loc

#Defining function to load data & save image from the url
def download_prep(pid,url,web,cat):
    '''
    input: unique product id, url of the image , Flipkart/Myntra data classification, produt category
    output: In case of problem with URL/process - errored product id is displayed. Otherwise, the loaded data is saved to respective folder location.
    '''
    try:
        url = url.strip(' ""')
        loc = './Data/'+str(web)+'/'+str(cat)+'/'+str(pid)+'.jpg'
        prod_id = generate_filename(loc)
        response=requests.get(url, stream=True, allow_redirects=True)
        open(prod_id,'wb').write(response.content)
    except:
        print(pid)
        errored_id.append(pid)
```

```python
# Data loading
errored_id = []
start_time1 = time.time()
for i in category:
    start_time = time.time()
    print("Category : ", i)
    cat_values = my_data[my_data['category_type'] == i]
    cat_values.apply(lambda x: download_prep(x['pid'],x['url_img'],x['Web'],i), axis = 1 )
    print("Time Taken    ", (time.time() - start_time) / 60 , " minutes")

print("Total Time taken  ", (time.time() - start_time1) / 60, " minutes")
print("completed")
```
