```python
#Load the Yolo results on the Flipkart Dataset
import pandas as pd
flipkart = pd.read_csv("./flipkart_results.csv")
flipkart.head()
```

```python
#Generalize the categories
Lower ='|'.join(['Bottom','Shorts','Pants'])
Upper = '|'.join(['Hoodie','Jackets','Tops'])
Full = '|'.join(['Jumpsuit','Cloak','Dress'])
flipkart.loc[flipkart['category'].str.contains(Lower),"actual_cat"] = "Lower"
flipkart.loc[flipkart['category'].str.contains(Upper),"actual_cat"] = "Upper"
flipkart.loc[flipkart['category'].str.contains(Full),"actual_cat"] = "Full"
flipkart.head()
```

```python
flipkart['category'].unique()
```

```python
#Setting the logic to get the images whose object detection score is greater than 0.70
def category_value(x,y):
    ls = x.strip('[]').replace("'","").split(', ')
    y = y.strip('[]').replace("'","").split(', ')  
    y = [float(i) for i in y if i != ""]
    if len(set(ls)) == 1:
        return list(set(ls))[0]
    else:
        gt_seven = sum(map(lambda a: a >= 0.70, y))
        if gt_seven == 1:   
            for i,cat in enumerate(ls):
                if y[i] > 0.70:
                    return cat
                else:
                    return "none"
        else:
            maxies = max(y)
            if maxies >= 0.70:
                ind = y.index(maxies)
                return ls[ind]
            else:
                return "none"
flipkart['pred_cat1']=flipkart.apply(lambda x: category_value(x['pred_cat'],x['pred_con']) , axis=1)
'''
def category_value(x,y):
    ls = x.strip('[]').replace("'","").split(', ')
    y = y.strip('[]').replace("'","").split(', ')  
    y = [float(i) for i in y if i != ""]
    if len(set(ls)) == 1:
        return list(set(ls))[0]
    else:
        maxies = max(y)
        ind = y.index(maxies)
        return ls[ind]
flipkart['pred_cat1']=flipkart.apply(lambda x: category_value(x['pred_cat'],x['pred_con']) , axis=1)
'''
```

```python
#Encoding the data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(['Lower', 'Upper', 'Full','none'])
```

```python
flipkart1 = flipkart[flipkart['category'] != "Mixed"]
```

```python
le.transform(flipkart1['actual_cat'])
```

```python
flipkart1['act_code'] = le.transform(flipkart1['actual_cat'])
flipkart1['pre_code'] = le.transform(flipkart1['pred_cat1'])
```

```python
flipkart1["Accuracy"] = flipkart1['act_code'] - flipkart1['pre_code']
flipkart1
```

```python
#Confusion Matrix
true_pos_upper = len(flipkart1[(flipkart1["Accuracy"]==0) & (flipkart1['actual_cat'] == 'Upper')])
true_pos_lower = len(flipkart1[(flipkart1["Accuracy"]==0) & (flipkart1['actual_cat'] == 'Lower')])
true_pos_full = len(flipkart1[(flipkart1["Accuracy"]==0) & (flipkart1['actual_cat'] == 'Full')])

upper_lower = len(flipkart1[(flipkart1["pred_cat1"]=='Lower') & (flipkart1['actual_cat'] == 'Upper')])
upper_full = len(flipkart1[(flipkart1["pred_cat1"]=='Full') & (flipkart1['actual_cat'] == 'Upper')])

lower_upper = len(flipkart1[(flipkart1["pred_cat1"]=='Upper') & (flipkart1['actual_cat'] == 'Lower')])
lower_full = len(flipkart1[(flipkart1["pred_cat1"]=='Full') & (flipkart1['actual_cat'] == 'Lower')])

full_upper = len(flipkart1[(flipkart1["pred_cat1"]=='Upper') & (flipkart1['actual_cat'] == 'Full')])
full_lower = len(flipkart1[(flipkart1["pred_cat1"]=='Lower') & (flipkart1['actual_cat'] == 'Full')])

Upper = ['Upper',true_pos_upper,upper_lower,upper_full]
Lower = ['Lower',lower_upper,true_pos_lower, lower_full]
Full = ['Full',full_upper,full_lower,true_pos_full]

confusion_matrix = pd.DataFrame(columns=['','Upper','Lower','Full'])
confusion_matrix.loc[0]=Upper
confusion_matrix.loc[1]=Lower
confusion_matrix.loc[2]=Full
confusion_matrix=confusion_matrix.set_index('')
confusion_matrix
```
