```python
# Loading Necessary Packages
import pandas as pd
import requests, shutil
import os
import time
```

```python
#Data Load
cloth_cate = pd.read_csv("./Data/list_category_cloth.txt",delimiter = "\t", header=0,error_bad_lines=False)
train_set = pd.read_csv("./Data/train.txt",delimiter = "\t",header = None)
train_bb = pd.read_csv("./Data/train_bbox.txt",header = None)
train_cate = pd.read_csv("./Data/train_cate.txt",header = None, names=['category_type_num'])
categories = pd.read_csv("./Data/category.csv")
test_set = pd.read_csv("./Data/test.txt",delimiter = "\t",header = None)
test_bb = pd.read_csv("./Data/test_bbox.txt",header = None)
test_cate = pd.read_csv("./Data/test_cate.txt",header = None, names=['category_type_num'])
```

```python
#Segmenting the category
cloth_category = list(cloth_cate['category_name  category_type'].map(lambda x: str(x)[:-1]).str.strip())
cloth_category = { cloth_category.index(i)+1 :  i for i in cloth_category }
train_cate['Values']=train_cate['category_type_num'].map(cloth_category)
test_cate['Values']=test_cate['category_type_num'].map(cloth_category)
categories_list = categories.apply(lambda x: x.dropna().tolist()).tolist()
# Filter data with regx
Shorts = '|'.join(categories_list[0])
Pants =  '|'.join(categories_list[1])
Mixed = '|'.join(categories_list[2])
Tops = '|'.join(categories_list[3])
Jackets = '|'.join(categories_list[4])
Jumpsuit = '|'.join(categories_list[5])
Bottom = '|'.join(categories_list[6])
Hoodie = '|'.join(categories_list[7])
Cloak = '|'.join(categories_list[8])
Dress = '|'.join(categories_list[9])
# Generalized category assignment
train_cate.loc[train_cate['Values'].str.contains(Shorts),"category_type"] = "Shorts"
train_cate.loc[train_cate['Values'].str.contains(Pants),"category_type"] = "Pants"
train_cate.loc[train_cate['Values'].str.contains(Tops),"category_type"] = "Tops"
train_cate.loc[train_cate['Values'].str.contains(Jackets),"category_type"] = "Jackets"
train_cate.loc[train_cate['Values'].str.contains(Bottom),"category_type"] = "Bottom"
train_cate.loc[train_cate['Values'].str.contains(Jumpsuit),"category_type"] = "Jumpsuit"
train_cate.loc[train_cate['Values'].str.contains(Hoodie),"category_type"] = "Hoodie"
train_cate.loc[train_cate['Values'].str.contains(Cloak),"category_type"] = "Cloak"
train_cate.loc[train_cate['Values'].str.contains(Dress),"category_type"] = "Dress"
test_cate.loc[test_cate['Values'].str.contains(Shorts),"category_type"] = "Shorts"
test_cate.loc[test_cate['Values'].str.contains(Pants),"category_type"] = "Pants"
test_cate.loc[test_cate['Values'].str.contains(Tops),"category_type"] = "Tops"
test_cate.loc[test_cate['Values'].str.contains(Jackets),"category_type"] = "Jackets"
test_cate.loc[test_cate['Values'].str.contains(Bottom),"category_type"] = "Bottom"
test_cate.loc[test_cate['Values'].str.contains(Jumpsuit),"category_type"] = "Jumpsuit"
test_cate.loc[test_cate['Values'].str.contains(Hoodie),"category_type"] = "Hoodie"
test_cate.loc[test_cate['Values'].str.contains(Cloak),"category_type"] = "Cloak"
test_cate.loc[test_cate['Values'].str.contains(Dress),"category_type"] = "Dress"
```

```python
#Making copy of the files 
train_data = train_cate.copy()
test_data = test_cate.copy()
train_data['images_loc']=train_set
train_data['bounding_box'] = train_bb
test_data['images_loc']=test_set
test_data['bounding_box'] = test_bb
```

```python
#Sampling of data
sample_data = train_data[train_data['Values'].str.contains(Bottom)].copy()
sample_data = sample_data.append(train_data[train_data['Values'] == 'Tee'].sample(frac=0.10))
sample_data = sample_data.append(train_data[train_data['category_type'] == 'Shorts'].sample(frac=0.19))
sample_data = sample_data.append(train_data[train_data['category_type'] == 'Pants'].sample(frac=0.18))
sample_data = sample_data.append(train_data[train_data['category_type'] == 'Dress'].sample(frac=0.20))
```

```python
#Saving the data to csv
sample_data.to_csv("sample_train.csv", index=False)
```
